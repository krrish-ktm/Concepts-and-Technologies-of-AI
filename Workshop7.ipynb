```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (mean_absolute_error, mean_squared_error, r2_score,accuracy_score, classification_report, precision_score , recall_score , f1_score)
import sklearn
sklearn.show_versions()
```

```python
df= pd.read_csv('/content/drive/MyDrive/Datasets/diabetes.csv')
df.head()
```

```python
X_bp = df.drop(columns=['BloodPressure', 'Outcome'])
y_bp = df['BloodPressure']
X_bp_train, X_bp_test, y_bp_train, y_bp_test = train_test_split(X_bp, y_bp, test_size=0.2, random_state=42)
lr_model = LinearRegression()
lr_model.fit(X_bp_train, y_bp_train)

y_bp_pred = lr_model.predict(X_bp_test)

mae = mean_absolute_error(y_bp_test, y_bp_pred)
mse = mean_squared_error(y_bp_test, y_bp_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_bp_test, y_bp_pred)

print("Regression Metrics:")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R2 Score: {r2}")
```

```python
X_outcome = df.drop(columns=['Outcome', 'BloodPressure'])
y_outcome = df['Outcome']
X_outcome_train, X_outcome_test, y_outcome_train, y_outcome_test = train_test_split(X_outcome, y_outcome, test_size=0.2, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_outcome_train, y_outcome_train)

y_outcome_pred_knn = knn.predict(X_outcome_test)
```

```python
print("KNN Evaluation:")
print(classification_report(y_outcome_test, y_outcome_pred_knn))
print(f"Accuracy: {accuracy_score(y_outcome_test, y_outcome_pred_knn)}")
print(f"Precision: {precision_score(y_outcome_test, y_outcome_pred_knn)}")
print(f"Recall: {recall_score(y_outcome_test, y_outcome_pred_knn)}")
print(f"F1 Score: {f1_score(y_outcome_test, y_outcome_pred_knn)}")
```
