#Task 1 - Sum of Nested Lists:

```python
def sum_nested_list(nested_list):

    total = 0
    for item in nested_list:
        if isinstance(item, list):
            total += sum_nested_list(item)
        else:
            total += item
    return total

# test the sample lisdt
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("The total sum of all numbers in the nested list is:", result)
```

#Task 2 - Generate All Permutations of a String:

```python
def generate_permutations(s):
    def backtrack(path, used, permutations):
        if len(path) == len(s):
            permutations.append(''.join(path))
            return
        for i in range(len(s)):

            if used[i] or (i > 0 and s[i] == s[i - 1] and not used[i - 1]):
                continue
            used[i] = True
            path.append(s[i])
            backtrack(path, used, permutations)
            path.pop()
            used[i] = False

    permutations = []
    sorted_chars = sorted(s)
    backtrack([], [False] * len(s), permutations)
    return permutations
# Teste the function with the example strings
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

#Task 3 - Directory Size Calculation:

```python
def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    return total_size
# Testinf the function with a sample directory structure
sample_directory = {
    "file1.txt": 700,
    "file2.txt": 1000,
    "subdir1": {
        "file3.txt": 500,
        "subdir2": {
            "file4.txt": 900,
            "file5.txt": 300,
        },
    },
    "file6.txt": 150,
}

result = calculate_directory_size(sample_directory)
print("The total size of the directory is:", result, "KB")
```

#Task 1 - Coin Change Problem:

```python
def min_coins(coins, amount):

    # Initialize the DP table with a large value (amount + 1, which is effectively infinity for this problem)
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test the function with coins = [1, 2, 5] and amount = 11
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"The minimum number of coins needed to make {amount} is: {result}")
```

#Task 2 - Longest Common Subsequence (LCS):

```python
def longest_common_subsequence(s1, s2):
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]

    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[len(s1)][len(s2)]

# Testing the function with example strings
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"The length of the LCS between '{s1}' and '{s2}' is: {result}")

print(f"The minimum number of coins needed to make {amount} is: {result}")
```

#Task 3 - 0/1 Knapsack Problem:

```python
def knapsack(weights, values, capacity):
    n = len(weights)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                # Maximize value: either include or exclude the current item
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    # The value at dp[n][capacity] is the maximum value achievable
    return dp[n][capacity]

# Test the function with example data
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"The maximum value achievable with capacity {capacity} is: {result}")
```
